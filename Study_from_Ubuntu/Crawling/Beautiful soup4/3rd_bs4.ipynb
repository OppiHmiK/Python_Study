```python
# -*- coding : utf-8 -*-
from bs4 import BeautifulSoup as bs
import urllib.request as req

url = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp'
```

```python
# NOTE : urlopen으로 데이터 불러오기
res = req.urlopen(url)
soup = bs(res, 'lxml')
soup
```

```python
title = soup.find('title').text
wf = soup.find('wf').text
print(title)
print(wf)
```
